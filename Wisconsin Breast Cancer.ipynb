```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('data.csv')
dataset.head(10)
```

```python
dataset_dummies = pd.get_dummies(data=dataset, columns=['diagnosis'],drop_first=True)
dataset_dummies.head(10)
```

```python
y = dataset_dummies.loc[:, ["diagnosis_M"]].values
dataset_dummies.drop(["diagnosis_M"], axis=1, inplace=True)
X = dataset_dummies.iloc[:, 1:].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
classifier.add(Dense(units = 24, kernel_initializer = 'uniform', activation = 'relu', input_dim = 30))
classifier.add(Dense(units = 15, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size = 4, epochs = 40)
```

```python
y_pred = classifier.predict(X_test)
from sklearn.metrics import confusion_matrix
y_pred_category = (y_pred > 0.3)
cm = confusion_matrix(y_test, y_pred_category)
cm
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
def build_cross_val_classifier():
    cv_classifier = Sequential()
    cv_classifier.add(Dense(units = 24, kernel_initializer = 'uniform', activation = 'relu', input_dim = 30))
    cv_classifier.add(Dense(units = 15, kernel_initializer = 'uniform', activation = 'relu'))
    cv_classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    cv_classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return cv_classifier
```

```python
cv_classifier = KerasClassifier(build_fn = build_cross_val_classifier, batch_size = 4, epochs = 100)
print(cv_classifier)
```

```python
accuracies = cross_val_score(estimator = cv_classifier, X=X_train, y=y_train, cv=10, n_jobs=1)
```

```python
average_accuracy = accuracies.mean()
std = accuracies.std()

print(f"acc: {average_accuracy} -- variance: {std}")
```
